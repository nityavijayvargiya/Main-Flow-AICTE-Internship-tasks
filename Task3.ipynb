```python
#Importing the Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/nityavijayvargiya/Main-Flow-AICTE-Internship-tasks/main/householdtask3.csv')
```

```python
data.head(10)
```

- year : Year in which the data was collected
- tot_hhs: Total number of households surveyed
- own: Number of household owners
- own_prop: Number of household owners proportion
- own_wm : Number of households where the woman owns the home
- wm_propprop: Proportion of women who are homeowners


```python
data.info()
```

```python
data.describe()
```

```python
#Creating a bar chart and a line chart using Matplotlib to visualize data from a Pandas DataFrame from above dataset

import matplotlib.pyplot as plt

# Create the bar chart
plt.bar(data['year'], data['tot_hhs'])
plt.xlabel('Year')
plt.ylabel('Total Households')
plt.title('Total Households by Year')
plt.show()

# Create the line chart
plt.plot(data['year'], data['own_prop'])
plt.xlabel('Year')
plt.ylabel('Homeownership Rate')
plt.title('Homeownership Rate by Year')
plt.show()
```

#Histograms


```python
#Distribution of age and income
data = data[['age', 'income']]

# Create the histogram
plt.hist(data, bins=10)

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Histogram of Age and Income")

# Show the plot
plt.show()
```

```python
#Understanding the central tendency, spread, and shape of the data.

# Select the most useful columns
cols = ['tot_hhs', 'own', 'own_wm', 'prop_hhs', 'age', 'size', 'income', 'expenditure', 'eqv_income', 'eqv_exp']

# Create the histogram
data[cols].hist(figsize=(10, 10), bins=10)

# Add title and labels
plt.title('Histogram of Most Useful Columns in Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Rotate x
```

```python
#proportion of households that own their homes
useful_data = data[['own_prop', 'own_wm_prop']]

# Create the histogram
plt.hist(useful_data, bins=10)
plt.xlabel('Percentage')
plt.ylabel('Frequency')
plt.title('Histogram of Home Ownership and Ownership with Mortgage')
plt.show()
```

```python
#visualizing the distribution of the data in the own column
data['own'].plot(kind='hist', bins=20, title='own')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

#Line Plot

```python
#Line plot shows the trend of total homeowners and the proportion of homeowners over time
data_top15 = data.head(15)

plt.figure(figsize=(10, 6))

plt.plot(data_top15['year'], data_top15['own'], label='Total Homeowners')
plt.plot(data_top15['year'], data_top15['own_prop'], label='Proportion of Homeowners')

plt.xlabel('Year')
plt.ylabel('Number of Households')
plt.title('Top 15: Homeowners vs. Proportion of Homeowners')

plt.legend()
plt.show()
```

```python
#Comparing the number of total household owners with the number of women household owners over time
plt.figure(figsize = (15, 7))
plt.plot(data['year'], data['own'], label = 'Total Household Owners')
plt.plot(data['year'], data['own_wm'], label = 'Women Household Owners')
plt.xlabel('Year')
plt.ylabel('Number of Households')
plt.title('Total and Women Household Owners in India from 2000 to 2012')
plt.legend()
plt.show()
```

```python
#Seeing how the number of women who own homes has changed over time.
data['own_wm'].plot(kind='line', figsize=(8, 4), title='own_wm')
plt.gca().spines[['top','right']].set_visible(False)
```

#Scatter Plot

```python
#Relationship between household size and income
plt.scatter(data['size'], data['income'])
plt.xlabel('size')
plt.ylabel('income')
plt.show()
```

```python
#Relationship between the year and the total number of households
data.plot(kind='scatter', x='year', y='tot_hhs', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

# Bar Charts

```python
#bar plot of the mean income and expenditure for each yea
data.groupby('year').mean()[['income', 'expenditure']].plot.bar()
```

#Pie Chart

```python
# Creating a pie chart of the proportion of households that own their homes
labels = ['Own', 'Rent']
sizes = [data['own_prop'].mean(), 100 - data['own_prop'].mean()]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.title('Proportion of Households that Own Their Homes')
plt.show()
```

#Box PLot

```python
# Create a list of lists containing the data for each boxplot
data = [
    data['size'],
    data['income'],
    data['expenditure']
]

# Set the labels for each boxplot
labels = ['Size', 'Income', 'Expenditure']

# Create the boxplot
fig, ax = plt.subplots(figsize=(10, 6))
ax.boxplot(data, labels=labels)

# Add a title and axis labels
ax.set_title('Boxplot of Size, Income, and Expenditure')
ax.set_xlabel('Variable')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
